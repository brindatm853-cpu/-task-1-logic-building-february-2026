1. smart parking lot management system

```python
def smart_parking_system(capacity, vehicle_logs):
    current_vehicles = 0
    peak_usage = 0
    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles
    if current_vehicles > capacity:
        status = "Exceeded Capacity!"
    else:
        status = "Available"
    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_system(capacity, vehicle_logs)
```

2. online food delivery time estimator

```python
def estimate_delivery_time(distance, traffic_level, weather):
    base_time = distance * 5
    if traffic_level.lower() == "low":
        traffic_delay = 5
    elif traffic_level.lower() == "medium":
        traffic_delay = 10
    elif traffic_level.lower() == "high":
        traffic_delay = 15
    else:
        traffic_delay = 0
    if weather.lower() == "clear":
        weather_delay = 0
    elif weather.lower() == "rainy":
        weather_delay = 10
    elif weather.lower() == "stormy":
        weather_delay = 20
    else:
        weather_delay = 0
    total_time = base_time + traffic_delay + weather_delay

    print("Estimated Delivery Time:", total_time, "minutes")
distance = 8
traffic_level = "High"
weather = "Rainy"
estimate_delivery_time(distance, traffic_level, weather)
```

3. movie theater seat occupancy analyser

```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100
    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    print("Occupancy:", int(occupancy_percentage), "%")
    print("Show Status:", status)
total_seats = 200
booked_seats = [1] * 150
seat_occupancy_analyzer(total_seats, booked_seats)
```

4. cloud server load classification system

```python
def classify_server_load(cpu_readings):
    if not cpu_readings:
        return "No readings provided"
    total_sum = sum(cpu_readings)
    count = len(cpu_readings)
    average_load = total_sum / count
    if average_load < 50:
        status = "Normal"
    elif 50 <= average_load < 80:
        status = "Warning"
    else:
        status = "Critical"
    print(f"Average CPU Load: {average_load:.0f}%")
    print(f"Server Status: {status}")
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

5. smart classroom resource usage monitor

```python
def monitor_resource_usage(usage_dict):
    """
    Identifies resources with usage exceeding 8 hours.
    """
    overused = []
    threshold = 8
    for resource, hours in usage_dict.items():
        if hours > threshold:
            overused.append(resource)
    energy_alert = "Yes" if overused else "No"
    overused_str = ", ".join(overused) if overused else "None"

    print(f"Overused Resources: {overused_str}")
    print(f"Energy Alert: {energy_alert}")
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resource_usage(resource_usage)
```

6. online event registration capacity controller

```python
def manage_event_registrations(capacity, total_attempts):
    confirmed = 0
    waitlisted = 0
    for i in range(1, total_attempts + 1):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    status = "Closed" if confirmed >= capacity else "Open"
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
event_capacity = 100
registrations_received = 105
manage_event_registrations(event_capacity, registrations_received)
```
