1. unique words in sentence


```python

sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

2. highest salary from employee data

```python

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]
print("Highest Salary:", highest_employee, "-", highest_salary)
```

3.find maximum and minimum values

```python

numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

4. count products above a price threshold

```python

prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

5. calculate attendance percentage

```python

attendance = ["P", "P", "A", "P", "P"]
present_count = 0
for status in attendance:
    if status == "P":
        present_count += 1
total_days = len(attendance)
percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

6. remove duplicate phone numbers

```python

phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

7. count character frequency

```python

text = "python"
char_count = {}
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1
print(char_count)
```

9.check if a key exists in a dictionary

```python

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
key_to_check = "Ravi"
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

10. calculate average marks

```python

marks = [70, 75, 80, 65]
total = sum(marks)
average = total / len(marks)
print("Average Marks:", average)
```
