1.user login check


```python

correct_username = "admin"
correct_password = "1234"
username = input("Enter username: ")
password = input("Enter password: ")
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2.pass/fail analyser


```python

marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("Total Students:", len(marks))
print("Number of Students Passed:", pass_count)
print("Number of Students Failed:", fail_count)
```

3.simple data cleaner

```python

names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

4.message length analyser

```python

messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("Status: Longer than 10 characters")
    else:
        print("Status: 10 or fewer characters")

    print("-----------------------------")
```

5.error message detector

```python

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = logs.count("ERROR")
print("Total Logs:", len(logs))
print("Number of ERROR entries:", error_count)
```
