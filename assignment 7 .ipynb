```python
 1.Smart Light Controller
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False
    def turn_on(self):
        self.is_on = True
    def turn_off(self):
        self.is_on = False
    def display_status(self):
        status = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

```python
2.Employee ID Card System
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
# Sample Run
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

```python
 3.Mobile Contact Record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
# Sample Run
contact = Contact("Anita", "9876543210")
contact.display_contact()
```

```python
 4.Product Price Tag Generator
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
# Sample Input
product_name = "Headphones"
price = 2499
# Create object
item = Product(product_name, price)
# Print formatted price tag
item.print_price_tag()
```

```python
 5.Movie Rating Display System
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")
# Sample Input
movie_name = "Inception"
rating = 4.8
# Create object
movie = Movie(movie_name, rating)
# Display movie details
movie.display_details()
```

```python
 6.Delivery Address Manager
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")
customer_name = "Suman"
address = "Hyderabad"
delivery = Delivery(customer_name, address)
delivery.print_details()
```
